```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def logistic_curve(t, A, K, B, theta, Q, C):
    return A + float(K - A) / (C + Q * np.e**(-B*t)**(1/theta))
```

```python
fig, ax = plt.subplots(3,2, figsize=(8,8), dpi=600)
#print ax
ax[0][0].hlines(0.5, 0, 1, label='birth rate', color='green',linestyle='-')
ax[0][0].hlines(0.5, 0, 1, label='death rate', color='blue',linestyle='--')
ax[0][0].set_ylim(0,1)
ax[0][0].get_xaxis().set_visible(False)
ax[0][0].get_yaxis().set_visible(False)
ax[0][0].text(0.58, 0.85, '"non-senescent"')

ax[0][1].plot(logistic_curve(np.linspace(-6,6,1000),0,1,8,1,1,1), label='death', color='blue',linestyle='--')
ax[0][1].plot(1-logistic_curve(np.linspace(-6,6,1000),0,1,8,1,1,1), label='birth', color='green',linestyle='-')
ax[0][1].get_xaxis().set_visible(False)
ax[0][1].get_yaxis().set_visible(False)
ax[0][1].text(600, 0.78, '"2-phase" or \n"rapid ageing"')

ax[1][0].plot(logistic_curve(np.linspace(-4,8,100),0,1,0.5,1,1,1), label='birth', color='green',linestyle='-')
ax[1][0].hlines(0.5,0,100, label='death', color='blue',linestyle='--')
ax[1][0].get_xaxis().set_visible(False)
ax[1][0].get_yaxis().set_visible(False)
ax[1][0].text(76, 0.74, '"negative\nageing"')

ax[1][1].plot(logistic_curve(np.linspace(-6,6,100),0,1,0.5,1,1,1), label='death', color='blue',linestyle='--')
ax[1][1].plot(1-logistic_curve(np.linspace(-6,6,100),0,1,0.5,1,1,1), label='birth', color='green',linestyle='-')
ax[1][1].get_xaxis().set_visible(False)
ax[1][1].get_yaxis().set_visible(False)
ax[1][1].text(78, 0.68, '"gradual\nageing"')

ax[2][0].plot(logistic_curve(np.linspace(-9,3,100),0,1,1.5,1,1,1), label='death', color='blue',linestyle='--')
ax[2][0].plot(1-logistic_curve(np.linspace(-3,9,100),0,1,1.5,1,1,1), label='birth', color='green',linestyle='-')
#ax[2][0].get_xaxis().set_visible(False)
ax[2][0].get_xaxis().set_ticks([])
ax[2][0].xaxis.set_ticklabels([])
ax[2][0].get_yaxis().set_visible(False)
ax[2][0].set_xlabel('age')
ax[2][0].text(34, 0.84, '"menopause"')

x = (0.25-norm.pdf(np.linspace(-10,10,100),0,2))
x[:30] *= 0.6
x[30:40] *= np.linspace(0.6,0.7,10)
x[40:50] *= np.linspace(0.7,0.8,10)
x[50:60] *= np.linspace(0.8,0.9,10)

ax[2][1].plot(norm.pdf(np.linspace(-10,10,100),0,3), label='birth', color='green',linestyle='-')
ax[2][1].plot(x-0.05, label='death', color='blue',linestyle='--')
#ax[2][1].get_xaxis().set_visible(False)
ax[2][1].xaxis.set_ticklabels([])
ax[2][1].get_xaxis().set_ticks([])
ax[2][1].get_yaxis().set_visible(False)
ax[2][1].set_xlabel('age')
ax[2][1].text(3, 0.18, '"reproductive peak"')

ax[0][0].legend(loc='lower left')
```
